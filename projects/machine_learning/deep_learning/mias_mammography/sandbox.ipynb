```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.patches import Circle
import numpy as np
import os
from PIL import Image
```

```python
plt.style.use('dark_background')
```

```python
mammo = pd.read_table('scan_file_data.txt', delimiter='\s', engine='python')
```

```python
mammo.columns = mammo.columns.str.lower()
```

```python
mammo.index.value_counts()
```

```python
mammo.iloc[243]
```

```python
# img = mpimg.imread('./all-mias/mdb132.pgm')
img = Image.open('./all-mias/mdb144.pgm')

# Create a figure. Equal aspect so circles look circular
fig, ax = plt.subplots(1)

fig.set_size_inches(12, 10)
ax.set_aspect('equal')

# Show the image
ax.imshow(img)
ax.set_ylim(bottom=0, top=1024)
print(ax.get_ylim()[0])

# create a circle to patch on the image
x = pd.to_numeric(mammo.iloc[145].x)
y = pd.to_numeric(mammo.iloc[145].y)
r = pd.to_numeric(mammo.iloc[145].radius)
circ = Circle((x,1024-y), r, fill=False)
ax.add_patch(circ)
print(x, y, r)

x1 = pd.to_numeric(mammo.iloc[146].x)
y1 = pd.to_numeric(mammo.iloc[146].y)
r1 = pd.to_numeric(mammo.iloc[146].radius)
circ1 = Circle((x1,1024-y1), r1, fill=False)
ax.add_patch(circ1)
print(x1, y1, r1)
# Show the image
plt.show()
```
