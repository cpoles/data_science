## Data Preparation

```python
# separating the predictors and the labels
housing = strat_train_set.drop('median_house_value', axis=1, inplace=False)
housing_labels = strat_train_set.median_house_value.copy() # deep copy
```

```python
# create a transformer for numerical data
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
rooms_idx, bedrooms_idx, population_idx, households_idx = 3, 4, 5, 6
```

```python
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no args or kwargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_idx] / X[:, households_idx]
        population_per_household = X[:, population_idx] / X[:, households_idx]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_idx] / X[:, rooms_idx]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

### Transformation Pipelines for Numerical and Categorical Attributes

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')), # fill out missing values with the median
    ('attribs_adder', CombinedAttributesAdder()), # add extra attributes
    ('std_scaler', StandardScaler()) # standardize attributes scale
    
])
```

```python
housing_num = housing.drop('ocean_proximity', axis=1)
```

```python
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']
```

```python
full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing)
```
