### Duplicate Elimination and Counting Frequencies

```python

# Using NumPy's unique function for list of frequencies
numbers = [random.randrange(1, 11) for _ in range(50)]

ns, frequencies = np.unique(numbers, return_counts=True)
```

```python
print(f'{list(zip(ns, frequencies))}')
```

### Survey Response Statistics

```python
# 1 awful - 5 excellent

responses = [1, 2, 5, 4, 3, 5, 2, 1, 3,3, 1, 4, 3, 3, 3, 2, 3, 3, 2, 5]
responses
```

```python
ratings, frequencies = np.unique(responses, return_counts=True)
print(f'{list(zip(ratings, frequencies))}')
```

```python
# measures of dispersion

print(f'Minimum: {min(responses)}')
print(f'Maximum: {max(responses)}')
print(f'Range: {np.ptp(responses)}')
print(f'Mean: {statistics.mean(responses)}')
print(f'Median: {statistics.median(responses)}')
print(f'Mode: {statistics.mode(responses)}')
print(f'Population Variance: {statistics.pvariance(responses)}')
print(f'Population Standard Deviation: {statistics.pstdev(responses):.2f}')
```

### Visualising Survey Response Statistics

```python
responses = [1, 2, 5, 4, 3, 5, 2, 1, 3,3, 1, 4, 3, 3, 3, 2, 3, 3, 2, 5]
responses
```

```python
# Ratings and Frequencies
ratings, frequencies = np.unique(responses, return_counts=True)

# Create initial bar plot
title = 'Student Cafeteria Ratings'
sns.set_style('whitegrid')
axes = sns.barplot(x=ratings, y=frequencies, palette='bright')

#set title and label axis
axes.set_title(title)
axes.set(xlabel='Ratings', ylabel='Frequency')
axes.set_ylim(top=max(frequencies) * 1.10)

# add frequency and percentage of ratings
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(responses)}'
    axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
```

### Coin Tossing

```python
tosses = [random.randrange(1, 3) for _ in range(20_000)]

# get toss and frequencies
toss, frequencies = np.unique(tosses, return_counts=True)
print(frequencies)
print(toss)

# Initial bar plot
title = 'Heads and Tails'
sns.set_style('whitegrid')
axes = sns.barplot(x=toss, y=frequencies)

axes.set_title(title)
axes.set(xlabel='Tosses', ylabel='Frequency')
axes.set_ylim(top=max(frequencies) * 1.10)

# add frequency and percentage of ratings
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(tosses)}'
    axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
```
